```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    'Date': ['2025-01-01', '2025-01-03', '2025-02-10', '2025-02-15', '2025-03-05', '2025-03-20'],
    'Product': ['Product A', 'Product B', 'Product A', 'Product C', 'Product B', 'Product C'],
    'Region': ['East', 'West', 'North', 'South', 'East', 'West'],
    'Quantity': [10, 5, 7, 12, 9, 15],
    'Unit Price': [50, 80, 50, 30, 80, 30]
}
```

```python
df = pd.DataFrame(data)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df['Total Sales'] = df['Quantity'] * df['Unit Price']
```

```python
total_sales_product = df.groupby('Product')['Total Sales'].sum()
print("Total Sales per Product:\n", total_sales_product)
```

```python
total_sales_product.plot(kind='bar', title='Total Sales per Product')
plt.xlabel('Product')
plt.ylabel('Total Sales')
plt.show()
```
